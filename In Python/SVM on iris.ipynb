```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.datasets import load_iris
# using seaborn for plotting
import seaborn as sns;
sns.set()
```

```python
# loading iris dataset from sklearn 
data = load_iris()
# convert dataset to pandas dataframe and assigning column names.
iris = pd.DataFrame(data=np.c_[data['data'],data['target']], columns=data['feature_names']+['target'])
type(iris)
iris
```

```python
# 2D plot
plt.scatter(iris['sepal length (cm)'],iris['sepal width (cm)'],c=iris['target'],cmap='autumn')
```

```python
# 3D plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(1, figsize=(16, 10))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(iris['sepal length (cm)'],iris['sepal width (cm)'], c=iris['target'],cmap=plt.cm.Set1, edgecolor='k', s=40)
```

```python
# split the training and test models
from sklearn.model_selection import train_test_split
X = iris[['sepal length (cm)','sepal width (cm)']]
y = iris.target
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 0,test_size = 0.3)
```

```python
# SVM
from sklearn.svm import SVC
svm = SVC(kernel='rbf',random_state=0,gamma=0.10,C=1.0)
svm.fit(X_train,y_train)
print('on training set:{}--on test set:{}'.format(svm.score(X_train,y_train),svm.score(X_test,y_test)))
```
