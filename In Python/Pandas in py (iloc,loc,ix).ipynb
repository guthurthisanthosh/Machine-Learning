```python
import pandas as pd
import random 
data = pd.read_csv('https://s3-eu-west-1.amazonaws.com/shanebucket/downloads/uk-500.csv')
# assigning id for employees randomly
data['id'] = [random.randint(0,1000) for x in range(data.shape[0])]
data
```

# 1.Selecting pandas data using iloc 

```python
# selecting columns
data.iloc[:,1]
data.iloc[0:2]
# selecting rows
data.iloc[0]
data.iloc[:,0:2]
```

```python
data.iloc[[0,1,2,3]]
data.iloc[[1,5,2,3,6],[5,2,6,3]]
```

# 2. Selecting pandas using 'loc'

```python
# 2(a) : Label based
# setting first_name as label and selecting data
data.set_index('first_name', inplace=True)
data.head(5)
```

for selecting using row value, that row needs to be the index of that dataset

```python
# select specific record from dataset using emp name
data.loc['Evan'] # single employee
data.loc[['Evan','France']] # multiple employees
data.loc[['Evan','France'],['city','county']] # select multiple employees with specific columns
data.loc[['Evan','France','Erasmo'],'city':'phone2']
```

Note that in the last example, data.loc[487] (the row with index value 487) is not equal to data.iloc[487] (the 487th row in the data). The index of the DataFrame can be out of numeric order, and/or a string or multi-value.

```python
# 2 (b) : Boolean/logical indexing using .loc
data.loc[data['first_name'] == 'Erasmo','city':'email'] # select's columns between city and email 
data.loc[data['first_name'] == 'Erasmo',['city','email']] # select's only city, email columns
```

# 3. Select Pandas data using ix 

1. it is a hybrid of .loc and .iloc
2. .loc can be replaced .ix when string is passed to it.
        eg:  data.ix[['Andrade']] == data.loc[['Andrade']]
3. loc can be replaced with .ix when int is passed to it.
        eg: data.ix[[33]] == data.iloc[[33]]

```python
# change values from dataFrame using .loc
# changes first_name to Evan if id > 0
data.loc[data['id'] > 0, 'first_name'] = "Evan"
data
```
