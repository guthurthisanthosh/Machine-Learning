```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
data = pd.read_csv('F:\\Linear Regression\\insurance.csv')
```

```python
data.head()
```

```python
x = data.drop('charges', axis=1)
x
```

```python
y = data.charges
y
```

```python
data_dummies = pd.get_dummies(data[['sex','smoker','region']])
data_dummies
```

```python
x = x.drop(['sex','smoker','region'], axis=1)
x
```

```python
x = pd.concat([data_dummies,x],axis=1)
x
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
```

```python
# r square
model.score(x,y)
```

```python
pd.Series(model.coef_)
```

```python
x.columns
```

```python
pd.concat([pd.Series(x.columns),pd.Series(model.coef_)], axis = 1)
```
